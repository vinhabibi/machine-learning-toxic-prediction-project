### Using `.loc` for Label-Based Indexing

`df.loc` is primarily used for selection by label (row and column names).

**1. Selecting rows by index label:**

```python
# Select the row with index label 0 (the first row)
display(df.loc[0])

# Select multiple rows by a list of index labels
display(df.loc[[0, 5, 10]])
```

**2. Selecting rows based on a boolean condition:**

```python
# Select all rows where 'Gender' is 'Female'
display(df.loc[df['Gender'] == 'Female'].head())
```

**3. Selecting specific columns by label:**

```python
# Select the 'City' column
display(df.loc[:, 'City'].head())

# Select multiple columns
display(df.loc[:, ['City', 'Sales', 'Product line']].head())
```

**4. Selecting both rows and columns:**

```python
# Select 'Sales' and 'Rating' columns for the first 5 rows
display(df.loc[0:4, ['Sales', 'Rating']])

# Select 'Sales' and 'Rating' columns for rows where 'Customer type' is 'Member'
display(df.loc[df['Customer type'] == 'Member', ['Sales', 'Rating']].head())
```

### 1. Distribution of Products by 'Product line'
This visualization shows the frequency of each product category in the dataset. It helps understand which product lines are most common.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(y='Product line', hue='Gender', data=df, order=df['Product line'].value_counts().index)
plt.title('Distribution of Products by Product Line and Gender')
plt.xlabel('Number of Sales Transactions')
plt.ylabel('Product Line')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(y='Product line', data=df, order=df['Product line'].value_counts().index)
plt.title('Distribution of Products by Product Line')
plt.xlabel('Number of transactions')
plt.ylabel('Product Line')
plt.show()
```
